```python
import os
from pathlib import Path
base = Path(r'Z:')
data_dir = base.joinpath("/Projects/LowdosePET/PiB_post022021/raw_data")
```

```python
for folder in data_dir.iterdir():
    counter_norm = 0
    counter_calib = 0
    counter_lm = 0
    print(folder.name)
    if folder.is_dir():
        for subf in folder.iterdir():
            
            if subf.is_dir():
                if subf.name == 'umap':
                    print(subf.name)
                else:
                    print('!!!!!!!!!!!!!!')
            if 'NORM' in subf.name:
                counter_norm += 1
            if 'CALIBRATION' in subf.name:
                counter_calib += 1
            if 'LISTMODE' in subf.name:
                counter_lm += 1
    
    print(counter_norm, counter_calib, counter_lm)
    print('\n')
```
